```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset from a CSV file
# Cargar el conjunto de datos desde un archivo CSV
df = pd.read_csv('UNdata_rail_lines.csv')

# Filter the dataframe to only include countries with less than 5000 km of total rail lines
# Filtrar el dataframe para incluir solo los países con menos de 5000 km de líneas ferroviarias en total
df = df[df['Value'] < 5000]

# Create a dataframe for rows where the Year is 1995
# Crear un dataframe con solo las filas donde el Año es 1995
df_country_1995 = df[df['Year'] == 1995]

# Create a dataframe for rows where the Year is 2019
# Crear un dataframe con solo las filas donde el Año es 2019
df_country_2019 = df[df['Year'] == 2019]

# Extract the list of countries that have data for both years (1995 and 2019)
# Extraer la lista de países que tienen datos tanto para los años 1995 como para 2019
country_1995 = df_country_1995["Country or Area"]   # Series
country_2019 = df_country_2019["Country or Area"]
countries = sorted(list(set(country_1995).intersection(set(country_2019))))

# Highlight the countries with the highest or lowest absolute increase in rail line length
# Resaltar los países con el aumento absoluto más alto o bajo en la longitud de las líneas ferroviarias
colors = {
    # Countries with a steep increasing slope: Korea and Israel
    # Países con una pendiente de aumento pronunciado: Corea e Israel
    "Korea": "green",
    "Israel": "green",

    # Countries with a steep decreasing slope: Latvia and Dem. Rep. Congo
    # Países con una pendiente de disminución pronunciada: Letonia y Rep. Dem. Congo
    "Latvia": "red",
    "Dem. Rep. Congo": "red"
}

# Create a plot with a specific figure size
# Crear un gráfico con un tamaño de figura específico
fig, ax = plt.subplots(1, figsize=(10,10), facecolor='lightgrey')
ax.set_facecolor('white')

# Loop through each country and plot its data
# Recorrer cada país y graficar sus datos
for i,v in enumerate(countries):
    cntry = df[df['Country or Area'] == v] # Filter the data for a single country
    # Plot the data for each country with specific formatting
    # Graficar los datos de cada país con un formato específico
    plt.plot(cntry.Year, cntry.Value,
             color = colors[v] if (v in colors) else "grey", # Color based on the country's increase
             lw=2.5, # Line width
             marker='o', markersize=5) # Marker settings

    # Adjust label positioning for overlapping country names (to make them readable)
    # Ajustar la posición de las etiquetas para evitar superposición de nombres de países (hacerlas legibles)
    push_down_left = 0
    if v in ['Bulgaria','Israel']:
        push_down_left = 100
    elif v in ['Tunisia','Vietnam']:
        push_down_left = 50
    elif v in ['North Macedonia']:
        push_down_left = -50  # Move up by 50

    push_down_right = 0
    if v in ['Bulgaria', 'Slovak Republic']:
        push_down_right = 100
    elif v in ['Greece','Vietnam']:
        push_down_right = 50

    # Label the starting point of the country's data (near the first year in the dataset)
    # Etiquetar el punto de inicio de los datos del país (cerca del primer año en el conjunto de datos)
    plt.text(cntry.Year.values[1]-0.6,
             cntry.Value.values[1] - push_down_left,
             v, ha='right')  # Horizontal alignment to the right

    # Label the ending point of the country's data (near the last year in the dataset)
    # Etiquetar el punto final de los datos del país (cerca del último año en el conjunto de datos)
    plt.text(cntry.Year.values[0]+0.6,
             cntry.Value.values[0] - push_down_right,
             v)

# Set the limits for the x-axis and define specific x-ticks
# Establecer los límites para el eje x y definir los ticks específicos del eje x
plt.xlim(1987,2027)
plt.xticks([1995, 2019])

# Label the y-axis (km) with specific font size
# Etiquetar el eje y (km) con un tamaño de fuente específico
plt.ylabel('km', fontsize='14')

# Get the current y-ticks and format them by replacing 1000 with 'k'
# Obtener los ticks actuales del eje y y formatearlos reemplazando 1000 con 'k'
yticks = plt.yticks()[0]

# [1:-1] removes the first and last y-ticks to avoid cluttering
# [1:-1] elimina los primeros y últimos ticks del eje y para evitar desorden
plt.yticks(yticks[1:-1],
           [f'{i/1000}k' for i in yticks[1:-1]])

# Add vertical grid lines with specific styling
# Agregar líneas de cuadrícula vertical con un estilo específico
ax.xaxis.grid(color='black', linestyle='solid',
              which='both', alpha=0.9)

# Add horizontal grid lines with specific styling
# Agregar líneas de cuadrícula horizontales con un estilo específico
ax.yaxis.grid(color='black', linestyle='dashed',
              which='both', alpha=0.1)

# Remove the spines (edges) of the plot for a cleaner look
# Eliminar los bordes del gráfico para una apariencia más limpia
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.spines['top'].set_visible(False)

# Set the title with a specific font
# Establecer el título con una fuente específica
font1 = {'family':'serif','color':'black','size':20}
plt.title('Rail lines (total route)\n', loc='center',
          fontdict=font1)

# Add a footer text with the data source and author info
# Agregar un texto de pie de página con la fuente de los datos y la información del autor
plt.figtext(0.02, 0.02,
            "Data: World Development Indicators | The World Bank   Viz: @roannav",
            ha="left",  # horizontalalignment
            fontsize=10)

# Save the plot as an image file
# Guardar el gráfico como un archivo de imagen
plt.savefig('day05_slope.png')  # savefig must happen before show()
#plt.show()  # Uncomment this line to display the plot in a window
```
