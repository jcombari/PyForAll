```python
import pandas as pd    # Import pandas library to work with dataframes / Importar la librería pandas para trabajar con dataframes
from joypy import joyplot    # Import the joyplot function from the joypy library to create ridgeline plots / Importar la función joyplot de la librería joypy para crear gráficos de ridgeline

# https://matplotlib.org/stable/tutorials/colors/colormaps.html
from matplotlib import cm    # Import colormaps from matplotlib to apply color gradients to the plot / Importar colormaps de matplotlib para aplicar gradientes de color al gráfico

import matplotlib.pyplot as plt    # Import matplotlib to customize and display the plot / Importar matplotlib para personalizar y mostrar el gráfico

df1 = pd.read_csv("SanBernardino_PM2_5_A.csv")    # Read the first CSV file into a dataframe / Leer el primer archivo CSV en un dataframe
df2 = pd.read_csv("SanBernardino_PM2_5_B.csv")    # Read the second CSV file into a dataframe / Leer el segundo archivo CSV en un dataframe
df = pd.concat([df1, df2])    # Concatenate the two dataframes into one / Concatenar los dos dataframes en uno solo

# make a ridgeline plot of pm2.5 levels in Central San Bernardino Mountains 
# during 2021.

#print(df.head())    # Optionally print the first few rows to preview the data / Imprimir opcionalmente las primeras filas para ver los datos
#print(df.info())    # Optionally print information about the dataframe structure / Imprimir opcionalmente la información sobre la estructura del dataframe
num_rows = df.shape[0]    # Get the number of rows in the dataframe / Obtener el número de filas en el dataframe

# You can't have a negative number of Particulate Matter (PM 2.5).
# < 0 values don't make sense, so remove them.
df = df[df['Value']>=0]    # Filter the dataframe to keep only rows where 'Value' is greater than or equal to 0 / Filtrar el dataframe para mantener solo las filas donde 'Value' es mayor o igual a 0
print(f"{num_rows - df.shape[0]} rows were deleted")    # Print how many rows were deleted due to negative values / Imprimir cuántas filas fueron eliminadas debido a valores negativos

# Extract the month part from the "Date Time" column.
# Use regular expression to save the 1st digit and optional 2nd digit 
# and discard the rest of the "Date Time".
month = df["Date Time"].astype(str).replace(r'^(\d)(\d?)/(.*)$', r'\g<1>\g<2>', 
    regex=True)    # Use regular expression to extract the month from the 'Date Time' column and store it in a new variable / Usar una expresión regular para extraer el mes de la columna 'Date Time' y guardarlo en una nueva variable

# add a column "Month" to the dataframe
df["Month"] = month    # Add a new column 'Month' to the dataframe with the extracted month values / Añadir una nueva columna 'Month' al dataframe con los valores del mes extraídos

# want ridgeline plot to sort the Month categories by number; not alphabetically
# so convert the values in Month from string to int
df = df.astype({'Month': 'int'})    # Convert the 'Month' column from string to integer type to sort correctly in the plot / Convertir la columna 'Month' de cadena a tipo entero para ordenar correctamente en el gráfico

#split it by months; a new density plot for each month.
fig, ax = joyplot(df, by='Month', column='Value',
    # other colormaps: bone, Purples, cividis, cool, twilight, twilight_shifted
    colormap=cm.twilight_shifted, fade=True,
    range_style='own',  # horizontal line under the ridgeline is only visible
                        # where the density is non-zero; 
                        # it doesn't span the whole x-axis
    tails=0,            # do not extend the tails for aesthetics / no extender las colas por estética
    figsize=(10, 6)    # dimensions of the figure in inches / dimensiones de la figura en pulgadas
)    # Create the ridgeline plot grouped by 'Month' and based on 'Value', with the specified aesthetics / Crear el gráfico de ridgeline agrupado por 'Month' y basado en 'Value', con la estética especificada

plt.xlabel("PM 2.5")    # Add label to the x-axis / Añadir etiqueta al eje x

# plt.annotate():
# first arg: text to display
# xy: location being annotated
# xytext: location of the text

plt.annotate('"Peak fire" in San Bernardino, CA\non 6/28/2021?', 
             xy=(175, 0.5), xytext=(205, 0.55),
             arrowprops=dict(facecolor='black', shrink=0.05))    # Add annotation for the peak fire on June 28, 2021 / Añadir anotación para el incendio de pico el 28 de junio de 2021

plt.annotate('Unknown cause', 
             xy=(280, 0.35), xytext=(205, 0.4),
             arrowprops=dict(facecolor='black', shrink=0.05))    # Add annotation for the unknown cause at a different point / Añadir anotación para la causa desconocida en un punto diferente

#move main title up
plt.subplots_adjust(top=.82)    # Adjust the position of the plot title to move it upwards / Ajustar la posición del título del gráfico para moverlo hacia arriba
plt.suptitle('Distribution of PM 2.5 Levels in Central San\nBernardino Mountains (CA) for Each Month in 2021', fontsize=20)    # Add a main title to the plot / Añadir un título principal al gráfico

plt.subplots_adjust(left=.08)    # Adjust the left margin of the plot / Ajustar el margen izquierdo del gráfico
plt.subplots_adjust(bottom=.15)    # Adjust the bottom margin of the plot / Ajustar el margen inferior del gráfico

# description text
plt.text(75,
    0.75,
    "Fine particulate matter (PM 2.5) is an air pollutant.  At high levels,\nit makes the air appear hazy.  High exposure in individuals can\naffect lung function and worsen asthma and heart disease.\nPM 2.5 is caused by vehicle exhausts, burning of fuels, fires,\npower plants, volcanic eruptions and tobacco smoke.",
    fontsize=12)    # Add descriptive text explaining the significance of PM 2.5 levels / Añadir texto descriptivo explicando la importancia de los niveles de PM 2.5

# footer (credits)
plt.text(0,
    -0.2,
    "Data: South Coast AQM",
    ha="center",  # horizontalalignment / alineación horizontal
    fontsize=10)    # Add footer text with data source information / Añadir texto en el pie de página con la información de la fuente de datos

plt.savefig('day08_mountains.png')  # savefig must happen before show()    # Save the plot as an image file / Guardar el gráfico como un archivo de imagen
#plt.show()    # Optionally display the plot (currently commented out) / Opcionalmente mostrar el gráfico (actualmente comentado)
```
