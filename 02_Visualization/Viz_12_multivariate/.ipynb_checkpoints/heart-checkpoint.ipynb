### **English** 🇺🇸

Hey there! 👋 This code is all about exploring heart disease data 💔 from a dataset, making it super easy to understand! Here's what it does:

1. **Data Cleaning**: It prepares the data to make sure everything is good to go 🚀 (like converting Yes/No answers into numbers like 1/0, and mapping age categories into numbers! 🧑‍🦳➡️👶).
   
2. **Sampling**: It grabs a smaller and balanced sample from the dataset (150 patients with heart disease and 150 without! ⚖️) so it’s easier to work with and analyze! 📉

3. **Exploring Data**: It checks out some cool info like how many people have heart disease ❤️‍🩹, what their age groups look like, how they feel mentally 🧠, and even how much they smoke 🚬. 

4. **Visualization**: It creates a **pairplot** 📊 that shows relationships between **Age**, **BMI**, and **Smoking** 🔥—highlighted by the **Heart Disease** target 💔 so you can easily spot trends and patterns!

Basically, it’s like creating a cool map 🗺️ of the data and understanding how different factors could be connected to heart disease! 🏥

### **Español** 🇪🇸

¡Hola! 👋 Este código tiene como objetivo explorar los datos sobre enfermedades cardíacas 💔 de un conjunto de datos y hacerlo super fácil de entender. ¡Aquí te cuento qué hace!

1. **Limpieza de Datos**: Prepara los datos para asegurarse de que todo esté listo para usar 🚀 (como convertir las respuestas de Sí/No en números como 1/0 y asignar números a las categorías de edad 🧑‍🦳➡️👶).

2. **Muestreo**: Toma una muestra más pequeña y equilibrada del conjunto de datos (¡150 personas con enfermedad cardíaca y 150 sin! ⚖️), para que sea más fácil de analizar y trabajar con ella. 📉

3. **Exploración de Datos**: Revisa datos geniales como cuántas personas tienen enfermedades cardíacas ❤️‍🩹, qué edades tienen, cómo están mentalmente 🧠 e incluso cuántos fuman 🚬.

4. **Visualización**: Crea un **pairplot** 📊 que muestra las relaciones entre **Edad**, **BMI** y **Fumar** 🔥, ¡destacando la enfermedad cardíaca 💔 para que puedas ver fácilmente las tendencias y patrones!

Básicamente, ¡es como crear un mapa genial 🗺️ de los datos y entender cómo diferentes factores podrían estar conectados con las enfermedades cardíacas! 🏥

```python
import matplotlib.pyplot as plt  # Importing the matplotlib library for plotting (importando la librería matplotlib para crear gráficos)
import pandas as pd  # Importing pandas library for data manipulation (importando la librería pandas para manipular datos)
import seaborn as sns  # Importing seaborn library for statistical visualizations (importando la librería seaborn para visualizaciones estadísticas)

# options are darkgrid, whitegrid, dark, white, or ticks (las opciones son darkgrid, whitegrid, dark, white o ticks)
#sns.set_style("ticks")  # Set the style of the plot (establecer el estilo del gráfico)

# many options such as bright, husl, Set2, dark, deep, etc. (muchas opciones como bright, husl, Set2, dark, deep, etc.)
#sns.set_palette("Set2")  # Set the color palette of the plot (establecer la paleta de colores del gráfico)

# Use set_theme() to set style and palette at the same time (usar set_theme() para establecer el estilo y la paleta al mismo tiempo)
#sns.set_theme(style="ticks")  # Set the theme to "ticks" (establecer el tema como "ticks")

df = pd.read_csv("heart_2020_cleaned.csv")  # Load the dataset from a CSV file (cargar el conjunto de datos desde un archivo CSV)

# There is no missing data already (ya no hay datos faltantes)
#df = df.dropna()  # drop rows where at least 1 element is missing (eg NaN) (eliminar filas donde al menos un elemento esté faltante (ejemplo NaN))

print(df.shape)  # Print the shape of the dataframe (imprimir las dimensiones del dataframe)

print(df.head())  # Print the first few rows of the dataframe (imprimir las primeras filas del dataframe)
print(df.columns)  # Print the column names (imprimir los nombres de las columnas)
print(df.describe())  # Print summary statistics for numerical columns (imprimir estadísticas descriptivas para las columnas numéricas)
print(df.info())  # Print information about the dataframe, including data types (imprimir información sobre el dataframe, incluyendo los tipos de datos)

# Target: HeartDisease: can be 'Yes' or 'No' (Objetivo: HeartDisease: puede ser 'Yes' o 'No')
print(df.HeartDisease.unique())  # Print unique values for the 'HeartDisease' column (imprimir los valores únicos para la columna 'HeartDisease')
print(df.HeartDisease.value_counts())  # Print the count of each value in 'HeartDisease' (imprimir la cantidad de cada valor en 'HeartDisease')

print(df.AgeCategory.value_counts())  # Print the count of each value in 'AgeCategory' (imprimir la cantidad de cada valor en 'AgeCategory')
print(len(df.AgeCategory.value_counts()))   # 13 (13 diferentes categorías de edad)

# Create a dictionary to map age categories to numbers (crear un diccionario para mapear las categorías de edad a números)
age_group = { 
    "18-24": 1, "25-29": 2, "30-34": 3, "35-39": 4, "40-44": 5, 
    "45-49": 6, "50-54": 7, "55-59": 8, "60-64": 9, "65-69": 10, 
    "70-74": 11, "75-79": 12, "80 or older": 13
}
df['AgeGroup'] = df['AgeCategory'].apply(lambda x: age_group[x])  # Map the AgeCategory column to AgeGroup (mapear la columna AgeCategory a AgeGroup)

# Print the value counts for the 'MentalHealth' column (imprimir la cantidad de cada valor en la columna 'MentalHealth')
print("df.MentalHealth.value_counts()")
print(df.MentalHealth.value_counts())
#plt.hist(df.MentalHealth, bins=30)  # Uncomment to create a histogram of 'MentalHealth' (descomentar para crear un histograma de 'MentalHealth')

df['Smoking'] = df['Smoking'].apply(lambda x: 1 if x=='Yes' else 0)  # Convert 'Smoking' from Yes/No to 1/0 (convertir 'Smoking' de Yes/No a 1/0)

print("df.Smoking.value_counts()")  # Print the value counts for the 'Smoking' column (imprimir la cantidad de cada valor en la columna 'Smoking')
print(df.Smoking.value_counts())  # Print the counts (imprimir las cantidades)
#plt.hist(df.Smoking)  # Uncomment to create a histogram of 'Smoking' (descomentar para crear un histograma de 'Smoking')

# Do you have serious difficulty walking or climbing stairs? Yes or No (¿Tienes serias dificultades para caminar o subir escaleras? Sí o No)
df['DiffWalking'] = df['DiffWalking'].apply(lambda x: 1 if x=='Yes' else 0)  # Convert 'DiffWalking' from Yes/No to 1/0 (convertir 'DiffWalking' de Yes/No a 1/0)

print("df.DiffWalking.value_counts()")  # Print the value counts for the 'DiffWalking' column (imprimir la cantidad de cada valor en la columna 'DiffWalking')
print(df.DiffWalking.value_counts())  # Print the counts (imprimir las cantidades)
#plt.hist(df.DiffWalking)  # Uncomment to create a histogram of 'DiffWalking' (descomentar para crear un histograma de 'DiffWalking')

# The dataset is huge: 319795 rows, (El conjunto de datos es enorme: 319795 filas,)
# so I'll take a smaller sample size. (así que tomaré una muestra más pequeña.)
# And have even number of HeartDisease 'No' and 'Yes' (Y tener un número equilibrado de 'HeartDisease' 'No' y 'Yes')
# so just 150 of each. (así que solo tomaré 150 de cada uno.)
# 
# The target HeartDisease has mostly 'No' and much less 'Yes' (El objetivo HeartDisease tiene principalmente 'No' y mucho menos 'Yes')
# No     292422 (No     292422)
# Yes     27373 (Yes     27373)

groupbyHeartDisease = df.groupby('HeartDisease')  # Group the dataframe by 'HeartDisease' (Agrupar el dataframe por 'HeartDisease')
df2 = groupbyHeartDisease.apply(lambda x: x.sample(n=150, random_state=1, replace=True))  # Sample 150 rows from each group (tomar 150 filas de cada grupo)

# Print information about the sampled dataframe (imprimir información sobre el dataframe muestreado)
print("\n\n")
print(df2.shape)  # Print the shape of the new sampled dataframe (imprimir las dimensiones del nuevo dataframe muestreado)

print(df2.head())  # Print the first few rows of the sampled dataframe (imprimir las primeras filas del dataframe muestreado)
print(df2.columns)  # Print the column names of the sampled dataframe (imprimir los nombres de las columnas del dataframe muestreado)
print(df2.describe())  # Print summary statistics for numerical columns (imprimir estadísticas descriptivas para las columnas numéricas)
print(df2.info())  # Print information about the sampled dataframe (imprimir información sobre el dataframe muestreado)
print(df2.HeartDisease.value_counts())  # Print the count of 'HeartDisease' values in the sampled dataframe (imprimir la cantidad de valores de 'HeartDisease' en el dataframe muestreado)
print(df2.AgeGroup.value_counts())  # Print the count of 'AgeGroup' values in the sampled dataframe (imprimir la cantidad de valores de 'AgeGroup' en el dataframe muestreado)

# Numerical data: (Datos numéricos:)
# MentalHealth: for how many days during the past 30 days was your mental (MentalHealth: ¿cuántos días durante los últimos 30 días estuvo tu salud mental no bien?)
# health not good? (0-30 days) (salud no fue buena? 0-30 días)
# 0 means the best mental health (0 significa la mejor salud mental)
# I will use BMI, SleepTime, and AgeGroup (Usaré BMI, SleepTime y AgeGroup)

sns.pairplot(df2, hue="HeartDisease",  # Create a pairplot to explore relationships between variables (crear un pairplot para explorar las relaciones entre variables)
    #kind='kde',  # Uncomment to change the plot type to KDE (descomentar para cambiar el tipo de gráfico a KDE)
    vars=['AgeGroup', 'BMI', 'Smoking'],  # Variables to include in the pairplot (Variables a incluir en el pairplot)
    #vars=['AgeGroup', 'BMI', 'SleepTime', 'MentalHealth', 'Smoking', 'DiffWalking'],  # Uncomment to include more variables (descomentar para incluir más variables)
    plot_kws=dict(alpha=0.5)  # Set transparency for the points in the plot (establecer transparencia para los puntos en el gráfico)
)

#plt.subplots_adjust(top=.82)  # Uncomment to adjust the plot title position (descomentar para ajustar la posición del título del gráfico)
#plt.suptitle('Distribution of PM 2.5 Levels in Central San\nBernardino Mountains (CA) for Each Month in 2021', fontsize=20)  # Add a main title (agregar un título principal)
#plt.subplots_adjust(bottom =.15)  # Uncomment to adjust the position of the footer (descomentar para ajustar la posición del pie de página)
#plt.text(0, 0,  # Add text at the bottom of the plot (agregar texto en la parte inferior del gráfico)
#    "Data: https://www.kaggle.com/datasets/kamilpytlak/personal-key-indicators-of-heart-disease   Viz: @roannav",
#    ha="left",  # horizontalalignment (alineación horizontal)
#    fontsize=10)  # Set font size for the footer text (establecer tamaño de fuente para el texto del pie de página)

plt.savefig('heart.png')  # Save the plot as a PNG file (guardar el gráfico como un archivo PNG)
```
