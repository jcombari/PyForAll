```python
# Importing necessary libraries
import itertools
import pandas as pd
from sklearn.datasets import load_iris
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Step 1: Load the Iris dataset from sklearn
iris = load_iris(as_frame=True)  # Load as a DataFrame
data = iris.frame  # Contains features and target
```

```python
# Step 2: Preview the dataset
print("Dataset Preview:")
print(data.head())
```

```python
# Step 3: Generate feature combinations
# Selecting only numerical columns for combinations
numerical_cols = iris.feature_names  # List of feature names
# Creating all possible combinations of features (2 at a time)
combinations = list(itertools.combinations(numerical_cols, 2))
combinations
```

```python
# Step 4: Print combinations and compute sample metrics
for combo in combinations:
    print(f"\nAnalyzing combination: {combo}")
    subset = data[list(combo)]
    print(f"Summary statistics for {combo}:")
    print(subset.describe())

# Step 5: Visualization (optional, for your video)
# Adding the target column for better visualization
data['species'] = iris.target
```

```python
# Plotting pairplots for the feature combinations
for combo in combinations:
    sns.pairplot(data, vars=combo, hue='species', palette='bright')
    plt.title(f"Feature combination: {combo}")
    plt.show()
```
