### üí° Ejemplo:
Usaremos el dataset de acciones financieras `yfinance` para calcular la volatilidad m√≥vil (desviaci√≥n est√°ndar) de un activo. El resultado es un an√°lisis din√°mico que te ayuda a tomar decisiones basadas en datos reales.

```python
import yfinance as yf
import numpy as np
import matplotlib.pyplot as plt

# Descargar datos hist√≥ricos
stock_data = yf.download(['MSFT'], start='2022-01-01', end='2022-12-31', group_by='ticker', interval='1d')

# Revisar las columnas disponibles
print("Available columns:", stock_data.columns)

# Selecci√≥n de precios de cierre (ajustando para el MultiIndex)
if ('MSFT', 'Close') in stock_data.columns:
    close_prices = stock_data[('MSFT', 'Close')]
else:
    raise KeyError("Column ('MSFT', 'Close') not found in the dataset.")

# Calcular la desviaci√≥n est√°ndar con rolling().apply()
rolling_volatility = close_prices.rolling(window=10).apply(lambda x: np.std(x), raw=True)

# Graficar los resultados
plt.figure(figsize=(10, 6))
plt.plot(close_prices, label='MSFT Close Price')
plt.plot(rolling_volatility, label='10-Day Rolling Volatility', color='orange')
plt.title('MSFT Prices and Rolling Volatility')
plt.xlabel('Date')
plt.ylabel('Price / Volatility')
plt.legend()
plt.grid()
plt.show()
```
