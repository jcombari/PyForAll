```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Generate a range of dates covering the entire year of 2025, with one entry per day
dates = pd.date_range(start="2025-01-01", end="2025-12-31", freq="D")

# Create a DataFrame using the generated date range
data = pd.DataFrame({"date": dates})

# Extract specific date components using the .dt accessor
# Extract the day of the week (as names) and flag weekends (Saturday and Sunday)
data["day_of_week"] = data["date"].dt.day_name()  # Day name (e.g., Monday, Sunday)
data["month"] = data["date"].dt.month  # Month as an integer (1 = January, 12 = December)
data["is_weekend"] = data["date"].dt.dayofweek >= 5  # Flag for weekends (Saturday/Sunday)

# Count the number of weekend days (Saturdays and Sundays) in each month
weekend_counts = data[data["is_weekend"]].groupby("month")["is_weekend"].count()

# Define month names in English for the plot labels
month_names = ["January", "February", "March", "April", "May", "June", 
               "July", "August", "September", "October", "November", "December"]

# Plot the number of weekend days in each month
plt.figure(figsize=(10, 6))
plt.bar(month_names, weekend_counts.values, color="skyblue", edgecolor="black")
plt.title("Number of Weekend Days per Month in 2025", fontsize=14)
plt.xlabel("Month", fontsize=12)
plt.ylabel("Number of Weekend Days", fontsize=12)
plt.xticks(rotation=45)  # Rotate labels for better readability
plt.grid(axis="y", linestyle="--", alpha=0.7)  # Add horizontal grid lines for better readability
plt.tight_layout()  # Adjust layout to ensure nothing is cropped
plt.show()
```
