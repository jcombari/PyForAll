# üêçüêº Python-Pandas: M√°s all√° de los gr√°ficos comunes con hist y m√∫ltiples ejes

## ‚ùì ¬øC√≥mo puedes visualizar varias distribuciones de datos sin sobrecargar el gr√°fico?
üôÉ Usando hist y m√∫ltiples ejes en Pandas para una exploraci√≥n de datos avanzada.

## üîß ¬øC√≥mo funciona?
Pandas permite crear histogramas directamente desde DataFrames, combinando varias columnas en un solo gr√°fico y utilizando m√∫ltiples ejes para superponer l√≠neas o barras sin necesidad de bibliotecas externas.

## üîé ¬øPor qu√© importa?
Visualizar m√∫ltiples distribuciones en un solo gr√°fico ayuda a detectar patrones ocultos, valores at√≠picos y tendencias en los datos sin depender de Matplotlib o Seaborn.

## ‚ú® Ejemplo real: An√°lisis de rendimiento de veh√≠culos üöó
Imagina que est√°s explorando un conjunto de datos de autos y quieres analizar la relaci√≥n entre la potencia del motor (hp), la aceleraci√≥n (acceleration) y el consumo de combustible (mpg).

## ‚öôÔ∏è Impacto en el negocio:
‚Ä¢ üìâ Optimizaci√≥n de consumo de combustible en flotas.  
‚Ä¢ ‚ö° Evaluaci√≥n del rendimiento de autos el√©ctricos frente a combusti√≥n.

## üìä ¬øQu√© hacemos en el c√≥digo?
‚Ä¢ Creamos un histograma con varias columnas (mpg, hp, acceleration).  
‚Ä¢ Superponemos una l√≠nea de densidad en un eje secundario.  
‚Ä¢ Todo esto usando solo Pandas.

## üí≠ Si pudieras analizar cualquier conjunto de datos con este m√©todo, ¬øqu√© informaci√≥n clave te gustar√≠a descubrir? üöÄüëá

```python
import pandas as pd  
import matplotlib.pyplot as plt  

# Load the built-in dataset from seaborn
from seaborn import load_dataset  
df = load_dataset("mpg").dropna()  # Remove missing values  

# Select relevant numerical columns
df = df[["mpg", "horsepower", "acceleration"]]  

# Rename columns for simplicity
df.columns = ["mpg", "hp", "acceleration"]  

# Create a histogram with multiple columns
ax = df.plot.hist(alpha=0.6, bins=20, figsize=(10, 6))  

# Overlay a density plot on a secondary y-axis
df["mpg"].plot(kind="kde", ax=ax, secondary_y=True, color="red", linestyle="--")  

# Add titles and labels
ax.set_xlabel("Value Range")  
ax.set_ylabel("Frequency")  
ax.right_ax.set_ylabel("Density")  
ax.set_title("Histogram and Density Plot of Vehicle Performance Data")  

# Show the plot
plt.show()  
```
