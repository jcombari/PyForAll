```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import yfinance as yf  # Import yfinance directly

# Define the time range for the analysis
start_date = "2023-01-01"
end_date = "2024-01-01"

# Fetch Tesla stock data using yfinance
tesla = yf.download('TSLA', start=start_date, end=end_date)

# Compute absolute difference between consecutive days
tesla['Price_Diff'] = tesla['Close'].diff()

# Compute percentage change between consecutive days
tesla['Pct_Change'] = tesla['Close'].pct_change() * 100  # Convert to percentage

# Create a figure to visualize trends
plt.figure(figsize=(12, 6))

# Plot Tesla closing price
plt.subplot(2, 1, 1)
plt.plot(tesla.index, tesla['Close'], label="Closing Price", color='blue')
plt.ylabel("Price (USD)")
plt.title("Tesla Stock Closing Price")
plt.legend()

# Plot percentage change
plt.subplot(2, 1, 2)
plt.plot(tesla.index, tesla['Pct_Change'], label="Daily % Change", color='red')
plt.axhline(0, linestyle='--', color='black', alpha=0.7)  # Add a reference line at 0%
plt.ylabel("Daily % Change")
plt.title("Tesla Stock Daily Percentage Change")
plt.legend()

# Show the plot
plt.tight_layout()
plt.show()
```
