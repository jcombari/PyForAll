- üåü Perfect for multivariate analysis or predictive models.

```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# Sample dataset: a DataFrame with users and their associated hashtags in lists
data = {
    "User": ["Alice", "Bob", "Charlie"],
    "Hashtags": [["#AI", "#ML"], ["#Python", "#DataScience", "#Pandas"], ["#DeepLearning", "#AI"]]
}

# Create a pandas DataFrame from the sample data
df = pd.DataFrame(data)

# Print the original dataset for reference
print("Original DataFrame:")
print(df)

# Use explode() to convert the list of hashtags into separate rows
df_exploded = df.explode("Hashtags")

# Print the transformed dataset
print("\nDataFrame after using explode():")
print(df_exploded)

# Group by 'Hashtags' and count occurrences to find the most popular hashtags
hashtag_counts = df_exploded["Hashtags"].value_counts()

# Print the counts of each hashtag
print("\nHashtag counts:")
print(hashtag_counts)

# Visualize the counts using a bar plot
plt.figure(figsize=(8, 5))
hashtag_counts.plot(kind="bar", color="skyblue")
plt.title("Most Popular Hashtags")
plt.xlabel("Hashtags")
plt.ylabel("Count")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```
