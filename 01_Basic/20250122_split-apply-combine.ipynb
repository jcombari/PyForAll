```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load a sample dataset (included in seaborn, also works with pandas)
from seaborn import load_dataset

# Load the "tips" dataset, which contains restaurant bill and tip data
data = load_dataset('tips')

# Preview the dataset to understand its structure
print(data.head())

# Group data by 'day' and 'time' and calculate summary statistics
grouped_data = data.groupby(['day', 'time'], observed=False).agg(
    total_bill_mean=('total_bill', 'mean'),  # Average total bill
    tip_mean=('tip', 'mean'),               # Average tip
    size_sum=('size', 'sum')                # Total number of people served
).reset_index()

# Display the grouped data
print(grouped_data)

# Create a bar chart to visualize the mean total bill and tip by day and time
fig, ax = plt.subplots(figsize=(10, 6))

# Bar positions and bar widths
x = np.arange(len(grouped_data['day'].astype(str) + " " + grouped_data['time'].astype(str)))
bar_width = 0.35

# Plot for mean total bill
ax.bar(x - bar_width/2, grouped_data['total_bill_mean'], bar_width, label='Avg Total Bill', color='skyblue')

# Plot for mean tip
ax.bar(x + bar_width/2, grouped_data['tip_mean'], bar_width, label='Avg Tip', color='orange')

# Add labels and title
ax.set_xlabel('Day & Time')
ax.set_ylabel('Value')
ax.set_title('Mean Total Bill and Tip by Day and Time')
ax.set_xticks(x)
ax.set_xticklabels((grouped_data['day'].astype(str) + " " + grouped_data['time'].astype(str)), rotation=45)
ax.legend()

# Show the plot
plt.tight_layout()
plt.show()


```
