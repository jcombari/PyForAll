```python
# Import necessary libraries
import pandas as pd
import seaborn as sns

# Load the Titanic dataset
titanic = sns.load_dataset('titanic')

# Display basic information about the dataset
print(titanic.info())  # View column data types and non-null counts
print(titanic.head())  # Preview the first 5 rows of the dataset

# Group passengers by class and calculate average fare using `itertuples`
# Prepare an empty dictionary to store results
class_fare = {}

# Iterate through rows using itertuples
for row in titanic.itertuples(index=False):
    # Access 'pclass' and 'fare' attributes
    if row.pclass not in class_fare:
        class_fare[row.pclass] = []
    class_fare[row.pclass].append(row.fare)

# Compute the average fare for each class
average_fare = {key: sum(values) / len(values) for key, values in class_fare.items()}

# Display the results
print("\nAverage Fare per Class:")
for pclass, avg_fare in average_fare.items():
    print(f"Class {pclass}: ${avg_fare:.2f}")

# Example output visualization
import matplotlib.pyplot as plt

# Plot the average fare per class
classes = list(average_fare.keys())
fares = list(average_fare.values())

plt.bar(classes, fares, color=['blue', 'green', 'red'])
plt.title('Average Fare by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Average Fare')
plt.xticks(classes, labels=['1st Class', '2nd Class', '3rd Class'])
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
