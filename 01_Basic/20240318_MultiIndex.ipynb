```python
# Import necessary libraries
import pandas as pd
import numpy as np

# Create a sample dataset
np.random.seed(42)  # For reproducibility
regions = ['North', 'South', 'East']
months = ['Jan', 'Feb', 'Mar']
categories = ['Sales', 'Cost', 'Profit']

# Create a MultiIndex for rows
row_index = pd.MultiIndex.from_product([regions, months], names=['Region', 'Month'])

# Create a MultiIndex for columns
col_index = pd.MultiIndex.from_product([categories, ['Mean', 'Total']], names=['Category', 'Metric'])

# Generate random data for the DataFrame
data = np.random.randint(100, 500, size=(len(row_index), len(col_index)))

# Create the DataFrame
df = pd.DataFrame(data, index=row_index, columns=col_index)

# Example Analysis: Aggregating data
region_summary = df.groupby(level='Region').sum()  # Total metrics by region
category_summary = df.groupby(axis=1, level='Category').sum()  # Total metrics by category

# Accessing specific data: Example query for 'North' region in 'Jan'
specific_data = df.loc[('North', 'Jan')]

# Print results
print("DataFrame with MultiIndex on Rows and Columns:")
print(df)
print("\nSummary by Region:")
print(region_summary)
print("\nSummary by Category:")
print(category_summary)
print("\nSpecific Data for 'North' Region in 'Jan':")
print(specific_data)
```
