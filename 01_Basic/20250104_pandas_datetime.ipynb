## üêçüêº Python-Pandas: Convierte cadenas en fechas con un truco r√°pido - pd.to_datetime

¬øLas fechas te complican? üìÖ A veces, analizarlas d√≠a a d√≠a es demasiado detallado, y hacerlo por a√±o resulta demasiado general. ¬°La soluci√≥n est√° en el punto medio: los meses! Aqu√≠ es donde pd.to_datetime y el agrupamiento por meses se convierten en tus mejores aliados.

### ‚ú® Por qu√© usar pd.to_datetime
- üîπ Convierte cadenas desordenadas en fechas √∫tiles.
- üîπ Simplifica el an√°lisis temporal a cualquier nivel: d√≠a, mes o a√±o.

### üíº Ejemplo pr√°ctico: Pasajeros del Titanic agrupados por mes
Imagina que est√°s analizando las fechas de registro de los pasajeros del Titanic. Primero, simularemos fechas aleatorias para cada pasajero, luego las agruparemos por mes para identificar patrones claros en el tiempo.

```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from seaborn import load_dataset

# Load the Titanic dataset
data = load_dataset('titanic')

# Step 1: Simulate random registration dates as strings
np.random.seed(42)  # Set seed for reproducibility
random_dates_strings = np.random.choice(
    pd.date_range(start='1912-01-01', end='1912-12-31').astype(str), size=len(data)
)
data['registration_date'] = random_dates_strings  # Add dates as strings

# Step 2: Convert registration_date strings to datetime format
data['registration_date'] = pd.to_datetime(data['registration_date'])

# Step 3: Extract the month from the registration_date
data['registration_month'] = data['registration_date'].dt.to_period('M')

# Step 4: Group data by registration_month and count passengers
monthly_grouped = data.groupby('registration_month').size().reset_index(name='passenger_count')

# Step 5: Plot the results
plt.figure(figsize=(12, 6))
plt.bar(monthly_grouped['registration_month'].astype(str), monthly_grouped['passenger_count'], color='skyblue')
plt.title('Passenger Registration by Month', fontsize=16)
plt.xlabel('Registration Month', fontsize=14)
plt.ylabel('Number of Passengers', fontsize=14)
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()

# Show the plot
plt.show()

```
