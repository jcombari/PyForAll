```python
import pandas as pd
import yfinance as yf

# Download hourly stock data for a specific symbol (e.g., Apple)
# We use the "period" and "interval" parameters for hourly data.
data = yf.download("AAPL", period="1mo", interval="1h")

# Inspect the first few rows of the dataset
print(data.head())

# Resampling: Convert hourly data to monthly means using 'ME' (Month-End)
# Resample the 'Close' price column to month-end frequency, calculating the mean
monthly_data = data['Close'].resample('ME').mean()

# Print the monthly resampled data
print("\nMonthly Resampled Data:")
print(monthly_data)

# Timezone normalization: Convert timezone to another timezone (e.g., New York)
# First, check if the index is already timezone-aware
if data.index.tz is None:
    data.index = data.index.tz_localize('UTC')  # Set timezone to UTC if not tz-aware

# Convert the timezone to New York time
data.index = data.index.tz_convert('America/New_York')

# Display data after timezone conversion
print("\nData with New York Timezone:")
print(data.head())

# Plot the original vs. resampled data
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.plot(data['Close'], label='Hourly Data', alpha=0.6)
plt.plot(monthly_data, label='Monthly Mean (Resampled)', marker='o', linewidth=2)
plt.title('Apple Stock Prices: Hourly vs. Monthly')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.grid()
plt.show()
```
