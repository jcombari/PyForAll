```python
# Importing necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Create a sample dataset
# This dataset represents sales data with regions, products, and years
data = {
    "Region": ["North", "North", "North", "South", "South", "South", "East", "East", "East"],
    "Product": ["A", "B", "C", "A", "B", "C", "A", "B", "C"],
    "Year": [2020, 2020, 2020, 2021, 2021, 2021, 2022, 2022, 2022],
    "Sales": [100, 200, 300, 400, 500, 600, 700, 800, 900]
}

# Step 2: Create a DataFrame from the dataset
df = pd.DataFrame(data)

# Display the original DataFrame
print("Original DataFrame:")
print(df)

# Step 3: Convert the DataFrame into a hierarchical structure using MultiIndex
df_multiindex = df.set_index(["Region", "Product", "Year"])

# Step 4: Sort the MultiIndex to optimize performance and avoid warnings
df_multiindex = df_multiindex.sort_index()

# Display the DataFrame with MultiIndex
print("\nDataFrame with MultiIndex:")
print(df_multiindex)

# Step 5: Querying sales for a specific region and product
# Example: Fetch sales for Region "North" and Product "A"
query_result = df_multiindex.loc["North", "A"]
print("\nQuery Result (Sales for North Region, Product A):")
print(query_result)

# Step 6: Pivot the data for visualization
# Rearrange data to display years as rows and regions/products as columns
df_plot = df.pivot_table(index="Year", columns=["Region", "Product"], values="Sales")

# Step 7: Plot the sales data
# Create a bar chart to visualize sales by year, region, and product
df_plot.plot(kind="bar", figsize=(10, 6), title="Sales by Year, Region, and Product")

# Add labels and format the plot
plt.ylabel("Sales")
plt.xlabel("Year")
plt.xticks(rotation=0)
plt.tight_layout()  # Adjust layout to avoid overlapping elements
plt.show()
```
