```python
# Import necessary libraries
import pandas as pd
import seaborn as sns
import zipfile
import io
import matplotlib.pyplot as plt

# Load the Titanic dataset from seaborn
titanic = sns.load_dataset('titanic')

# Save the dataset as a CSV inside a ZIP file (simulate data source)
buffer = io.BytesIO()
with zipfile.ZipFile(buffer, 'w', zipfile.ZIP_DEFLATED) as zf:
    zf.writestr("titanic.csv", titanic.to_csv(index=False))

# Reset the buffer's position
buffer.seek(0)

# Read the dataset directly from the ZIP file
with zipfile.ZipFile(buffer) as zf:
    with zf.open("titanic.csv") as file:
        df = pd.read_csv(file)

# Display the first few rows of the dataset
print("Dataset Preview:")
print(df.head())

# Grouping example: Analyzing survival rate by class and gender
grouped = df.groupby(['pclass', 'sex'])['survived'].mean()

# Print the grouped data
print("\nSurvival Rate by Class and Gender:")
print(grouped)

# Visualize the data
grouped.unstack().plot(kind='bar', stacked=True, figsize=(10, 6))
plt.title("Survival Rate by Class and Gender")
plt.ylabel("Survival Rate")
plt.xlabel("Passenger Class")
plt.legend(title="Gender")
plt.show()

```
