```python
# Importing libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load Titanic dataset
titanic = sns.load_dataset("titanic")

# Overview of the dataset
print("Dataset Overview:")
print(titanic.head())

# Converting specific columns to categorical types
titanic["class"] = pd.Categorical(titanic["class"], categories=["Third", "Second", "First"], ordered=True)
titanic["sex"] = pd.Categorical(titanic["sex"])
titanic["survived"] = pd.Categorical(titanic["survived"], categories=[0, 1], ordered=True)

# Memory usage before and after conversion
print("\nMemory Usage Before and After Conversion:")
print(titanic.info(memory_usage="deep"))

# Grouping by class, sex, and survival status
grouped = titanic.groupby(["class", "sex", "survived"], observed=False).size().reset_index(name="count")

# Display grouped data
print("\nGrouped Data:")
print(grouped)

# Preparing labels for the plot by concatenating class and gender
grouped["label"] = grouped["class"].astype(str) + " - " + grouped["sex"].astype(str)

# Plotting the data
plt.figure(figsize=(8, 6))
for survival_status in grouped["survived"].cat.categories:
    subset = grouped[grouped["survived"] == survival_status]
    plt.bar(subset["label"], subset["count"], label=f"Survived: {survival_status}")

plt.title("Survival Counts by Class and Gender")
plt.xlabel("Class and Gender")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.legend(title="Survival Status")
plt.tight_layout()
plt.show()
```
