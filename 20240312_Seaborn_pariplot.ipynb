```python
import seaborn as sns
import matplotlib.pyplot as plt

# Cargar el dataset de diamantes
data = sns.load_dataset('diamonds')

# Verificar las primeras filas del dataset para asegurarnos de que "cut" est presente
print(data.head())

# Seleccionar las variables que queremos analizar
variables = ['carat', 'depth', 'table', 'price']

# Verificar que "cut" es una columna vlida
if 'cut' in data.columns:
    # Crear un Pairplot con estas variables, utilizando 'cut' como hue
    sns.pairplot(data[variables + ['cut']], hue='cut', palette='coolwarm')

    # Ajustar el tamao del grfico
    plt.gcf().set_size_inches(12, 12)

    # Mostrar el grfico
    plt.show()
else:
    print("La columna 'cut' no est presente en el dataset.")
```

```python
# Import necessary libraries
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Load the 'mpg' dataset from seaborn
df = sns.load_dataset('mpg').dropna()  # Drop missing values

# Set up the figure for subplots
fig, axes = plt.subplots(3, 4, figsize=(20, 15))

# 1. Pairplot: Visualize pairwise relationships in the dataset
sns.pairplot(df[['mpg', 'horsepower', 'weight', 'acceleration']])

# 2. Correlation matrix: Visualize correlation between variables
sns.heatmap(df[['mpg', 'horsepower', 'weight', 'acceleration']].corr(), annot=True, cmap='coolwarm', ax=axes[0, 1])

# 3. 3D scatter plot (using 2D plot with color gradient for 3 variables)
sns.scatterplot(x='weight', y='horsepower', hue='mpg', data=df, palette='coolwarm', ax=axes[0, 2])

# 4. Multiple regression plot
sns.regplot(x='weight', y='mpg', data=df, ax=axes[0, 3])

# 5. Violin plot
sns.violinplot(x='origin', y='mpg', data=df, ax=axes[1, 0])

# 6. Heatmap of correlations (already done in 2)
# 7. Polynomial regression (fit a quadratic model)
sns.regplot(x='weight', y='mpg', data=df, order=2, ax=axes[1, 1])

# 8. Boxplot with categories
sns.boxplot(x='origin', y='mpg', data=df, ax=axes[1, 2])

# 9. Contour plot (using kdeplot)
sns.kdeplot(x='weight', y='horsepower', data=df, cmap='Blues', ax=axes[1, 3])

# 10. KDE (Kernel Density Estimation) plot
sns.kdeplot(df['mpg'], shade=True, ax=axes[2, 0])

# 11. Scatter plot with regression
sns.regplot(x='horsepower', y='mpg', data=df, ax=axes[2, 1])

# 12. Combined subplots (already done in multiple subplots above)
# Show all plots
plt.tight_layout()
plt.show()
```
