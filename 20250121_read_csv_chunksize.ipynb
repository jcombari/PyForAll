```python
import pandas as pd
import matplotlib.pyplot as plt
from dask.datasets import timeseries

# Generate a large dataset using Dask's timeseries
df = timeseries(start="2000-01-01", end="2000-12-31", freq="1s")
df = df.compute()  # Convert Dask dataframe to Pandas dataframe
df.to_csv("large_taxi_dataset.csv", index=False)  # Save locally as a CSV file

# Define the chunk size
chunk_size = 100000

# Initialize a list to store results
mean_x_values = []

# Process the dataset in chunks
for chunk in pd.read_csv("large_taxi_dataset.csv", chunksize=chunk_size):
    # Calculate mean of a specific column (example: 'x')
    mean_x = chunk['x'].mean()
    mean_x_values.append(mean_x)

# Combine results
overall_mean_x = sum(mean_x_values) / len(mean_x_values)

# Visualization
plt.figure(figsize=(10, 6))
plt.plot(mean_x_values, marker='o', label='Mean X per chunk')
plt.axhline(y=overall_mean_x, color='r', linestyle='--', label='Overall Mean X')
plt.title('Mean X Processed in Chunks')
plt.xlabel('Chunk Number')
plt.ylabel('Mean X')
plt.legend()
plt.grid()
plt.show()
```
